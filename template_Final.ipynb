```python
from scipy import stats
import numpy as np 
import matplotlib.pyplot as plt 
```

__Foi feito um _for_ para percorrer todas as variaveis.__

Ao utilizar-mos o _Jarque-bera_ iremos determinar quais as distribuições que passam o teste de normalidade.

__Neste teste o return terá dois valores :__

_jb-value_ , sendo este o valor estatístico.

_p-value_ , sendo este a probabilidade de se obter uma estatisica de teste igual ou mais extrema que aquela obtida.


Considerando o valor referência de 0.05 , se o _p-value_ for menor que este então iremos rejeitar esta distribuição. Caso seja maior considera-se uma distribuição normal, aceitando a mesma.

Ao analisarmos todas as distribuições testadas verificamos que o _p-value_ corresponde a 0.0 em todas. De acordo ao teste de _Jarque-Bera_ a distribuição testada nao é normal.

## Pergunta Bónus

```python
X = variavel_Independente
```

```python
y = variavel_Dependente
```

```python
import numpy as np
from numpy import array
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=0)
kf.get_n_splits(X)
i = 0
lista = []

for train_index, test_index in kf.split(X):
    
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    regressor = LinearRegression()  
    regressor.fit(X_train, y_train)

    y_pred = regressor.predict(X_test).round()
    
    mcc = matthews_corrcoef(y_test,y_pred)
    lista.insert(i,mcc)
    i = i + 1
    print(mcc)
```

O K-fold é o método que consiste em dividir o dataset em k partes, usando k-1 partes para treino e a parte restante para teste, fazendo isso k vezes. Em cada uma das k vezes, testa-se o modelo com um fold diferente calculando a métrica escolhida para avaliação do modelo. 

Após calcularmos, obtivemos 5 medidas porque o nosso dataset está dividido em 5 partes.

```python
avg_mcc = np.mean(lista)
print(avg_mcc)
```

Com o resultado das 5 medições, agora ao calcularmos a média teremos o verdadeiro score de como o modelo esta a generalizar
